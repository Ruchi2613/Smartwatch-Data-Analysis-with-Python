```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("/Users/ruchi/Downloads/Smartwatch-Data-Analysis-using-Python-main/dailyActivity_merged.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
# Changing datatype of ActivityDate
data["ActivityDate"] = pd.to_datetime(data["ActivityDate"], 
                                      format="%m/%d/%Y")
print(data.info())
```

```python
data["TotalMinutes"] = data["VeryActiveMinutes"] + data["FairlyActiveMinutes"] + data["LightlyActiveMinutes"] + data["SedentaryMinutes"]
print(data["TotalMinutes"].sample(5))
```

```python
figure = px.scatter(data_frame = data, x="Calories",
                    y="TotalSteps", size="VeryActiveMinutes", 
                    trendline="ols", 
                    title="Relationship between Calories & Total Steps")
figure.show()
```

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]
counts = data[["VeryActiveMinutes", "FairlyActiveMinutes", 
"LightlyActiveMinutes", "SedentaryMinutes"]]. mean()
colors = ['gold', "lightgreen", "pink,blue"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Total Active Minutes')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Sample data (Ensure columns exist)
data = pd.DataFrame({
    "Day": ["Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"],
    "VeryActiveMinutes": [30, 40, 25, 50, 35],
    "FairlyActiveMinutes": [20, 25, 15, 30, 22],
    "LightlyActiveMinutes": [50, 60, 55, 70, 65]
})

# Remove spaces from column names
data. columns = data.columns.str.strip()

# Fill NaN values if any
data.fillna(0, inplace=True)

# Plotting
fig = go.Figure()
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["VeryActiveMinutes"],
    name='Very Active',
    marker_color='purple'
))
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["FairlyActiveMinutes"],
    name='Fairly Active',
    marker_color='green'
))
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["LightlyActiveMinutes"],
    name='Lightly Active',
    marker_color='pink'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Sample data (Ensure columns exist)
data = pd.DataFrame({
    "Day": ["Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"],
    "VeryActiveMinutes": [30, 40, 25, 50, 35],
    "FairlyActiveMinutes": [20, 25, 15, 30, 22],
    "LightlyActiveMinutes": [50, 60, 55, 70, 65]
})

# Remove spaces from column names
data.columns = data.columns.str.strip()

# Fill NaN values if any
data.fillna(0, inplace=True)

# Now we use the correct column 'Day' instead of 'ActivityDate'
day = data["Day"].value_counts()
label = day.index
counts = data["VeryActiveMinutes"]  # Example: using 'VeryActiveMinutes' for counts
colors = ['gold', 'lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

# Pie chart for the 'Day' count and very active minutes
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Very Active Minutes by Day')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```
